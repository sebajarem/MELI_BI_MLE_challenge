## pruebas con Trie

```python
from pytrie import SortedStringTrie as Trie
```

```python
from pytrie import StringTrie

frases = list(ac.df_freq_frases.frase)
frases = list(ac.df_freq_frases_ult_salida.frase_salida)

t = StringTrie()

for f in frases:
    t[f] = f



```

```python
searches = ("when", "what")
```

```python
for search in searches:
    print('busqueda {}, da como resultado: {}'.format(search, t.values(prefix=search)[:5]))
```
